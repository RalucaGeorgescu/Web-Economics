#### Sample split is important

#Best random forest so far 
clf = RandomForestClassifier(n_estimators=100, max_features = 2, oob_score = True, random_state = 2, 
                             min_impurity_split= 0.002, min_samples_leaf = 10,                                                         
                             min_weight_fraction_leaf= 0,min_samples_split = 10,class_weight= 'balanced_subsample')

```python
#Feel free to change the parameters
clf = RandomForestClassifier(n_estimators=100, max_features = 8, oob_score = True, random_state = 2, 
                             min_impurity_split= 0.002, min_samples_leaf = 10, 
                             min_weight_fraction_leaf= 0,min_samples_split = 10,class_weight= 'balanced_subsample') #balanced_subsample
```

```python
%%time
clf.fit(X_train, y_train)
```

```python
clf_predict = clf.predict(X_test)
```

```python
clf.oob_score_
```

```python
for feature, importance in zip(X_train.columns,clf.feature_importances_):
    print (feature, ":", importance)
```

```python
acc_df = pd.DataFrame([clf_predict.transpose(),y_test], index=['Predict','Test']).transpose()
```
