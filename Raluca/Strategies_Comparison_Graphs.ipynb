## Comparing Strategies

```python
import matplotlib.pyplot as plt

labels = ['constant','random','linear','ORTB1','ORTB2']
strategies = [0,1,2,3,4]
clicks = [82,73,161,158,150]

plt.plot(strategies, clicks, 'g--o')
plt.xticks(strategies, labels)
plt.suptitle('Clicks')
plt.show()
```

```python
eCPC = [76.22,86.61,38.82,39.80,40.55]

plt.plot(strategies, eCPC, 'g--o')
plt.xticks(strategies, labels)
plt.suptitle('eCPC')
plt.show()
```

```python
impressions = [128601,77345,99328,63688,138024]

plt.plot(strategies, impressions, 'g--o')
plt.xticks(strategies, labels)
plt.suptitle('Impressions')
plt.show()
```

```python
CTR = [0.064,0.094,0.162,0.249,0.108]

plt.plot(strategies, CTR, 'g--o')
plt.xticks(strategies, labels)
plt.suptitle('CTR')
plt.show()
```
